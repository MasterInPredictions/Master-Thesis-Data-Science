```python
from sklearn.metrics import mean_squared_error
import statistics
```

```python
true_ = []
pred_ = []
pred_mlp = []

for i in range(6):
    true_.append(df.iloc[99:,i])
    
for j in range(-6,1):
    pred_.append(prediction_df.iloc[99:,j])
    
for j in range(6):
    pred_mlp.append(prediction_df_mlp.iloc[99:,j])
```

```python
mse_ = [[] for q in range(6)]
bias_ = [[] for q in range(6)]
variance_ = [[] for q in range(6)]
check_ = [[] for q in range(6)]

mse_mlp = [[] for q in range(6)]
bias_mlp = [[] for q in range(6)]
variance_mlp = [[] for q in range(6)]


for j in range(6):
    for i in range(42):
        if i == 0:
            mse_[j].append(mean_squared_error(true_[j][i:i+12],pred_[j][i:i+12]))
            mse_mlp[j].append(mean_squared_error(true_[j][i:i+12],pred_mlp[j][i:i+12]))
            
            variance_[j].append(statistics.variance((true_[j][i:i+12] - pred_[j][i:i+12])))
            variance_mlp[j].append(statistics.variance((true_[j][i:i+12] - pred_mlp[j][i:i+12])))
            
            bias_[j].append( (statistics.mean(true_[j][i:i+12]) - statistics.mean(pred_[j][i:i+12])) **2)
            bias_mlp[j].append( (statistics.mean(true_[j][i:i+12]) - statistics.mean(pred_mlp[j][i:i+12])) **2)
            
        
        if i == 41:
            break
        if i > 0:
            mse_[j].append(mean_squared_error(true_[j][i*12:(i+1) *12],pred_[j][i*12:(i+1) *12]))
            mse_mlp[j].append(mean_squared_error(true_[j][i*12:(i+1) *12],pred_mlp[j][i*12:(i+1) *12]))
            
            variance_[j].append(statistics.variance((true_[j][i*12:(i+1) *12] - pred_[j][i*12:(i+1) *12])))
            variance_mlp[j].append(statistics.variance((true_[j][i*12:(i+1) *12] - pred_mlp[j][i*12:(i+1) *12])))
            
            bias_[j].append( (statistics.mean(true_[j][i*12:(i+1) *12]) - statistics.mean(pred_[j][i*12:(i+1) *12]) ) **2)
            bias_mlp[j].append( (statistics.mean(true_[j][i*12:(i+1) *12]) - statistics.mean(pred_mlp[j][i*12:(i+1) *12]) ) **2)
         
            
```
